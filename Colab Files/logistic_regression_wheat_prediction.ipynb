```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LogisticRegression
```

```python
train_data = pd.read_csv('/content/wheet_train.csv')
tr = train_data.drop(['ID'],axis=1)
tr.head()
```

```python
y = tr['Type']
X = tr.drop(['Type'], axis=1)
```

```python
test_data = pd.read_csv('/content/wheet_test.csv')
test_data.head()
```

```python
tst = test_data.drop(['ID'],axis=1)
```

```python
log_reg = LogisticRegression()
log_reg = log_reg.fit(X,y)
```

```python
y_test = log_reg.predict(tst)
print(y_test)
```

```python
id = test_data['ID']
```

```python
import csv
submission = open("log_regr.csv","w")
sub_file = csv.writer(submission)
sub_file.writerow(['ID','Type'])
for i in range(0,len(y_test)):
    sub_file.writerow([str(id[i]),str(y_test[i])])
    
submission.close()
```
